```python
#import required libraries 
#import OpenCV library
import cv2
#import matplotlib library
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# cv2.cvtColor is an OpenCV function to convert images to different color spaces
def convertToRGB(img): 
    return cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
def detect_faces(f_cascade, colored_img, scaleFactor = 1.1):
 #just making a copy of image passed, so that passed image is not changed 
 img_copy = colored_img.copy()          
 
 #convert the test image to gray image as opencv face detector expects gray images
 gray = cv2.cvtColor(img_copy, cv2.COLOR_BGR2GRAY)          
 
 #let's detect multiscale (some images may be closer to camera than others) images
 faces = f_cascade.detectMultiScale(gray, scaleFactor=scaleFactor, minNeighbors=5);          
 
 #go over list of faces and draw them as rectangles on original colored img
 for (x, y, w, h) in faces:
      cv2.rectangle(img_copy, (x, y), (x+w, y+h), (0, 255, 0), 2)              
 
 return img_copy
```

## 作業一

```python
# load image
image = cv2.imread(r'data/test5.JPG')
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(gray_img, cmap = 'gray')
```

```python
# constuct detective class
haar_face_cascade = cv2.CascadeClassifier('data/haarcascade_frontalface_alt.xml')
faces = haar_face_cascade.detectMultiScale(gray_img, scaleFactor=1.08, minNeighbors=5)
len(faces)
```

```python
# plot faces
image_plot = image.copy()
for (x, y, w, h) in faces:
    cv2.rectangle(image_plot, (x,y), (x+w, y+h), (0,255,0), 5)
```

```python
plt.figure(figsize = (20,20))
plt.imshow(image_plot, cmap = 'gray')
```
