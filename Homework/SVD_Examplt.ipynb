```python
import numpy as np
import pandas as pd
import skimage.io as skio
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
class image_svd():
    def __init__(self):
        pass
        
    def load_img(self, fpath):
        self.imarray = skio.imread(fpath, as_grey = True)
    
    def show_img(self):
        plt.imshow(self.imarray)
    
    def svd_pic(self, k):
        U, s, V = np.linalg.svd(self.imarray)
        Uk = U[:, :k]
        Sk = np.diag(s[:k])
        Vk = V[:k, :]
        # recover the image
        imgMat_new = np.dot(np.dot(Uk,Sk), Vk)
        return imgMat_new
    
    def svd_pic_show(self, k):
        plt.imshow(self.svd_pic(k))
    
    def get_eigen_dim(self):
        U, s, V = np.linalg.svd(self.imarray)
        return U.shape
    eigen_dim = property(get_eigen_dim)
```
