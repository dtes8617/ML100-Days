### 用 iris (dataset.load_iris()) 資料嘗試跑 kmeans (可以測試不同的群數 , init 等)

```python
from sklearn import datasets

iris = datasets.load_iris()
X = iris.data
y = iris.target
```

### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
params = [(3, 'k-means++'), (3, 'random'),
          (5, 'k-means++'), (5, 'random'),
          (7, 'k-means++'), (7, 'random'),]

fig = plt.figure(figsize=(16,18))
for index, (groups, method) in enumerate(params):
    km = KMeans(groups, init = method)
    km.fit(X)
    km.labels_
    ax = fig.add_subplot(3, 2, index+1, projection='3d')
#     ax =Axes3D(fig, azim = -60)
    ax.scatter(X[:,0], X[:,1], X[:,2], c = km.labels_, edgecolor = 'k')
    ax.set_title('clusters: {}, init: {}'.format(groups, method))
```

```python
fig = plt.figure(figsize=(8,6))
ax =Axes3D(fig)
ax.scatter(X[:,0], X[:,1], X[:,2], c = y, edgecolor = 'k')
ax.set_title('Original dataset')
```
