# PCA 範例
* 以 PCA + 邏輯斯迴歸判斷手寫辨識資料集, 觀察不同 component 下正確率的變化

```python
# 載入套件
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.linear_model import SGDClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings("ignore")
```

```python
# 定義 PCA 與隨後的羅吉斯迴歸函數
# penalty 改為 L1, max_iter 改為 100
logistic = SGDClassifier(loss='log', penalty='l1', max_iter=100, tol=1e-5, random_state=0)
pca = PCA()
pipe = Pipeline(steps=[('pca', pca), ('logistic', logistic)])

# 載入手寫數字辨識集
digits = datasets.load_digits()
X_digits = digits.data
y_digits = digits.target
```

# 作業
* 將參數 penalty 改為 L1, max_iter 改為 100, 觀察 PCA 與羅吉斯迴歸做手寫辨識的效果

```python
# 執行 GridSearchCV 跑出最佳參數
params = {
    'pca__n_components': [4, 10, 20, 30, 40, 50, 64],
    'logistic__alpha': np.logspace(-4, 4, 5)
}
grid = GridSearchCV(pipe, params)
grid.fit(X_digits, y_digits)
```

```python
grid.best_params_
```

```python
# 繪製不同 components 的 PCA explained variance ratio
pca.fit(X_digits)

fig, (ax0, ax1) = plt.subplots(2, 1, figsize = (12,8), sharex = True)
ax0.plot(pca.explained_variance_ratio_)
ax0.set_ylabel('PCA explained varience')
ax0.axvline(x = grid.best_params_['pca__n_components'], linestyle = ':', label = 'n_components choosen')
ax0.legend(loc = 0, fontsize = 12)

# 繪製不同採樣點的 accuracy
result = pd.DataFrame(grid.cv_results_)
best_components = result.groupby('param_pca__n_components').apply(lambda x: x.nlargest(1, 'mean_test_score'))
best_components.plot(x = 'param_pca__n_components', y = 'mean_test_score', ax = ax1, yerr = 'std_test_score', legend = False)
ax1.set_ylabel('Classification accuracy (val)')
ax1.set_xlabel('n_components')
plt.tight_layout()
```
