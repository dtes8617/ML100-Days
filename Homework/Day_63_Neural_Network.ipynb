# sigmoid 函數圖形
* 講義中的 sigmoid 程式對照

```python
import matplotlib.pylab as plt 
import numpy as np 
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
x = np.arange(-8, 8, 0.1) 
f = 1 / (1 + np.exp(-x)) 
plt.plot(x, f) 
plt.xlabel('x') 
plt.ylabel('f(x)') 
plt.show()
```
